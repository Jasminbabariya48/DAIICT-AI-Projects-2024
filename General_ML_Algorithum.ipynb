# **CLASSIFICATION**

```python
import numpy as np
import pandas as pd
```

```python
#For more information about the dataset visit: https://archive.ics.uci.edu/dataset/451/breast+cancer+coimbra
cancer_dataset = pd.read_csv('/content/dataR2.csv')
```

```python
cancer_dataset.head(10)
```

```python
cancer_dataset.tail(10)
```

Statistic

```python
cancer_dataset.describe()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Histograms
cancer_dataset.hist(bins=10, figsize=(15, 10))
plt.show()

# Scatter plots
sns.pairplot(cancer_dataset)
plt.show()

# Correlation heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(cancer_dataset.corr(), annot=True, cmap='coolwarm', linewidths=0.5)
plt.show()
```

```python
x = cancer_dataset.iloc[:, 1:-1].values
y = cancer_dataset.iloc[:, -1].values
```

```python
y
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.25, random_state = 0)
```

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
x_train_scaler = sc.fit_transform(x_train)
x_test_scaler = sc.transform(x_test)
```

```python
x_train_scaler = pd.DataFrame(x_train_scaler)
x_test_scaler = pd.DataFrame(x_test_scaler)
```

Linear Regression

```python
#Write Code for Logistic Regression here
from sklearn.linear_model import LogisticRegression

classifier = LogisticRegression(random_state = 0)
classifier.fit(x_train_scaler, y_train)
y_pred = classifier.predict(x_test_scaler)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score,precision_score,recall_score

cm = confusion_matrix(y_test, y_pred)
print("Confusion matrix:",cm)
print("Accuracy of the model:",accuracy_score(y_test, y_pred))
print("Precision of the model:",precision_score(y_test, y_pred))
print("Recall of the model:",recall_score(y_test, y_pred))
```

SVM(Support Vector Machines)

```python
#Write Code for SVM here
from sklearn.svm import SVC

classifier = SVC(kernel = 'linear', random_state = 0)
classifier.fit(x_train_scaler, y_train)
y_pred1 = classifier.predict(x_test_scaler)
```

```python
cm = confusion_matrix(y_test, y_pred1)
print("Confusion matrix:",cm)
print("Accuracy of the model:",accuracy_score(y_test, y_pred1))
print("Precision of the model:",precision_score(y_test, y_pred1))
print("Recall of the model:",recall_score(y_test, y_pred1))
```

# **LINEAR REGRESSION**

```python
#For more information for the dataser refer: https://archive.ics.uci.edu/dataset/477/real+estate+valuation+data+set
real_estate_dataset = pd.read_excel("/content/Real estate valuation data set.xlsx")
```

```python
real_estate_dataset.tail(10)
```

Statistics

```python
real_estate_dataset.describe()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Histograms
real_estate_dataset.hist(bins=10, figsize=(15, 10))
plt.show()

# Scatter plots
sns.pairplot(real_estate_dataset)
plt.show()

# Correlation heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(real_estate_dataset.corr(), annot=True, cmap='coolwarm', linewidths=0.5)
plt.show()
```

```python
x = real_estate_dataset.iloc[:, :-1].values
y = real_estate_dataset.iloc[:, -1].values
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.25, random_state = 0)
```

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
x_train_scaler = sc.fit_transform(x_train)
x_test_scaler = sc.transform(x_test)
```

```python
#Code for Linear Regression here
from sklearn.linear_model import LinearRegression

regressor = LinearRegression()
regressor.fit(x_train_scaler, y_train)
y_pred2 = regressor.predict(x_test_scaler)
```

```python
from sklearn.metrics import r2_score

print("Score of the model",r2_score(y_test, y_pred2))
```
